## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

def calculate_subtotal(unit_price, qty):
    return unit_price * qty

df["subtotal"] = calculate_subtotal(df["unit_price"], df["qty"])
df.head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.





def shipping_price(unit_weight):
    if unit_weight >  50:
        return unit_weight * 7
    return unit_weight *  10  

df['shipping_price'] = df['unit_weight'].apply(shipping_price)
df.head()


print(df.columns)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%


def calculate_total_price(subtotal, shipping_price):
    sales_tax_rate = 0.0925
    tax_amount = subtotal * sales_tax_rate
    total_price = subtotal + tax_amount + shipping_price
    return total_price

df["total_price"] = df.apply(lambda row: calculate_total_price(row["subtotal"], row["shipping_price"]), axis = 1)
 
print(df.head())
```

```python
print(df[['first', 'last', 'total_price']].head())
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

def calculate_line_cost(unit_cost, qty, shipping_price):
    total_cost = (unit_cost * qty) + shipping_price
    return total_cost
df["line_cost"] = df.apply(lambda row: calculate_line_cost(row["unit_cost"], row["qty"], row["shipping_price"]), axis = 1)


```

```python
print(df[['first', 'last', 'line_cost']].head().round(2))
```

```python
print(df.columns)
```

```python
# Create a column for the profit of
# each line using line cost and line price
def calculate_profit (unit_cost, unit_price):
    profit = (unit_cost - unit_price)
    return profit

df["profit"] = df.apply(lambda row: calculate_profit(row["unit_cost"], row["unit_price"]), axis = 1)

```

```python
print(df[['first', 'last', 'profit']].head().round(2))
```
